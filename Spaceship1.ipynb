Vamos construir una variable categorica para agrupar edades en 5 categorias: Child (0 a 9), Teen (10 a 18), Young (19 a 29), Adult (30 a 40), (Senior) mayores de 41 y ver su relacion con transported

```python
def categorizar_edad(age):
    if pd.isnull(age):
        return "Unknown"
    elif age <= 9:
        return "Child"
    elif age <= 18:
        return "Teen"
    elif age <= 29:
        return "Young"
    elif age <= 40:
        return "Adult"
    else:
        return "Senior"

train_df["AgeGroup"] = train_df["Age"].apply(categorizar_edad)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=train_df, x="AgeGroup", hue="Transported", order=["Child", "Teen", "Young", "Adult", "Senior", "Unknown"])
plt.title("Transporte por grupo de edad")
plt.show()
```

```python
pd.crosstab(train_df["AgeGroup"], train_df["Transported"], normalize='index')\
  .round(2)\
  .sort_index()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Codificar variables categricas (One-Hot o Label Encoding)
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# Alinear columnas (por si difieren tras One-Hot)
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# Entrenar modelo
model = RandomForestClassifier(random_state=42)
scores = cross_val_score(model, X_train_encoded, y_train, cv=5, scoring='accuracy')
print("Accuracy promedio (CV):", scores.mean())

# Entrenar sobre todo el train
model.fit(X_train_encoded, y_train)

# Predicciones
predictions = model.predict(X_test_encoded)

```

Una forma sencilla es promediar las probabilidades que entregan ambos modelos y luego tomar la clase con la mayor probabilidad.

```python
from sklearn.ensemble import RandomForestClassifier

final_rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=2,
    random_state=42
)

final_rf.fit(X_train_encoded, y_train)
```

```python
# Obtener probabilidades (probabilidad de clase positiva "Transported")
rf_probs = final_rf.predict_proba(X_test_encoded)[:, 1]
xgb_probs = final_xgb.predict_proba(X_test_encoded)[:, 1]

# Promedio simple de probabilidades
ensemble_probs = (rf_probs + xgb_probs) / 2

# Convertir probabilidades a clases (umbral 0.5)
ensemble_preds = (ensemble_probs >= 0.5)

# Crear archivo submission
submission_ensemble = pd.DataFrame({
    'PassengerId': passenger_ids,
    'Transported': ensemble_preds
})

submission_ensemble.to_csv('submission_ensemble.csv', index=False)
print("Archivo submission_ensemble.csv creado con xito.")
```

```python
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Modelos base
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=2,
    random_state=42
)

xgb = XGBClassifier(
    colsample_bytree=0.7,
    learning_rate=0.01,
    max_depth=5,
    n_estimators=200,
    subsample=0.7,
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)

# Meta-modelo
meta_model = LogisticRegression()

# Crear el stacking
stacking_clf = StackingClassifier(
    estimators=[('rf', rf), ('xgb', xgb)],
    final_estimator=meta_model,
    cv=5,
    n_jobs=-1
)

# Evaluar con validacin cruzada
cv_scores = cross_val_score(stacking_clf, X_train_encoded, y_train, cv=5, scoring='accuracy')
print(f'Accuracy promedio (CV) stacking: {cv_scores.mean()}')

# Entrenar stacking final
stacking_clf.fit(X_train_encoded, y_train)

# Predecir en test
stacking_preds = stacking_clf.predict(X_test_encoded)

# Crear submission
submission_stacking = pd.DataFrame({
    'PassengerId': passenger_ids,
    'Transported': stacking_preds
})

submission_stacking.to_csv('submission_stacking.csv', index=False)
print("Archivo submission_stacking.csv creado.")
```
